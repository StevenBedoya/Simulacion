# Distribuciones Discretas

## Distribuci√≥n Binomial 


```python
%matplotlib inline 

import numpy as np
import scipy as sp
import math
import matplotlib.pyplot as plt
import scipy.integrate
n=10
b=[]
p=[]
bu=[]
acum=[]
#generador Aleatorio
def numalea():
    xo=7
    for i in range(n):
        mod=(223*xo+23)%829
        ui=float(mod)/829
        xo=mod
        p.append(ui)
    #print(p)
    return p


def binomial():
    p=numalea()
    for i in range(n):
        pr=(1-p[2])**(n-i)
        c=(math.factorial(n)/(math.factorial(i)*math.factorial(n-i)))*(p[2]**i)*pr
        b.append(c)
        #print(c)
    return b


def acumubino():
    g=0
    for i in range(n):
        g=b[i]+g
        acum.append(g)
    #print("Acumulada",acum)
    return acum
plt.subplot(2,1,1)


plt.plot(binomial(),label="Binomial",color="r",marker="o")
plt.plot(acumubino(), label="Acumulada")
plt.legend(loc="upper left")
plt.show()
plt.subplot(3,1,3)
def inverbino():
    p=numalea()
    for i in range(n):
        for j in range(n):
            if p[i]<acum[j]:
                bu.append(j)
                break
    print("Valores Binomial Inversa",bu)
    return bu
plt.plot(inverbino(),color="g",marker="x")
plt.show()
        
```

## Distribucion Geometrica

```python

ll=[]
acumg=[]
def geometrica():
    p=numalea()
    for i in range(n):
        l=((1-p[2])**(i-1))*p[2]
        ll.append(l)
    #print(ll)
    return ll

plt.subplot(2,1,1)
def acumugeome():
    h=0
    for i in range(n):
        h=ll[i]+h
        acumg.append(h)
    #print("Acumulada",acum)
    return acumg

plt.plot(geometrica(),label="Geometrica",color="r",marker="o")
plt.plot(acumugeome(), label="Acumulada")
plt.legend(loc="upper left")

plt.show()
plt.subplot(2,1,2)
bj=[]
def invergeo():
    p=numalea()
    for i in range(n):
        for j in range(n):
            if p[i]<acumg[j]:
                bj.append(j)
                break
    print("Valores Geometrica Inversa ",bj)
    return bj
plt.plot(invergeo(),marker="x", color="g")
plt.grid()
plt.show()
        
        
    
```

## Distribucion Uniforme discreta

```python

lo=[]
acumu=[]

def unidiscreta():
    p=numalea()
    for i in range(n):
        u=1/float(n)
        lo.append(u)
    #print(ll)
    return lo


plt.subplot(3,1,2)
def acumudis():
    hi=0
    for i in range(n):
        hi=lo[i]+hi
        acumu.append(hi)
    #print("Acumulada",acum)
    return acumu


plt.plot(unidiscreta(),label="Discreta",color="r",marker="o")
plt.plot(acumudis(), label="Acumulada")
plt.legend(loc="upper left")

plt.show()
plt.subplot(3,1,3)
bl=[]
def invergeo():
    p=numalea()
    for i in range(n):
        for j in range(n):
            if p[i]<acumu[j]:
                bl.append(j)
                break
    print("Valores Uniforme Inversa ",bl)
    return bl
plt.plot(invergeo(),marker="x", color="g")
plt.grid()
plt.show()
        
```
